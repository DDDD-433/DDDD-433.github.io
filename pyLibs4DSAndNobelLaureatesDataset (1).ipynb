# Analyzing Average Fuel Consumption and CO2 Emissions by Ship Type

In this section, we aim to answer the first research question:
**"What is the average fuel consumption and CO2 emissions for different ship types?"**

We will:
1. Group the data by `ship_type`.
2. Calculate the average fuel consumption and CO2 emissions for each ship type.
3. Visualize the results using bar plots for better understanding.


```python
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'ship_fuel_efficiency.csv'
data = pd.read_csv(file_path)

# Group the data by 'ship_type' and calculate average fuel consumption and CO2 emissions
ship_type_summary = data.groupby('ship_type')[['fuel_consumption', 'CO2_emissions']].mean().reset_index()

# Display summary table
print("Summary Statistics for Fuel Consumption and CO2 Emissions by Ship Type:")
display(ship_type_summary)

# Plot the average fuel consumption and CO2 emissions for different ship types
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Average Fuel Consumption by Ship Type
sns.barplot(x='ship_type', y='fuel_consumption', data=ship_type_summary, hue=None, palette='viridis', ax=ax[0])
ax[0].set_title('Average Fuel Consumption by Ship Type')
ax[0].set_xlabel('Ship Type')
ax[0].set_ylabel('Fuel Consumption (liters)')

# Average CO2 Emissions by Ship Type
sns.barplot(x='ship_type', y='CO2_emissions', data=ship_type_summary, hue=None, palette='plasma', ax=ax[1])
ax[1].set_title('Average CO2 Emissions by Ship Type')
ax[1].set_xlabel('Ship Type')
ax[1].set_ylabel('CO2 Emissions (kg)')

plt.tight_layout()
plt.show()

```

# Analyzing Fuel Consumption Across Different Routes

In this section, we aim to answer the second research question:
**"How does fuel consumption vary across different routes?"**

We will:
1. Group the data by `route_id` and calculate the average fuel consumption.
2. Sort the routes to identify those with the highest and lowest average fuel consumption.
3. Visualize the results using a horizontal bar plot for better readability.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by 'route_id' and calculate average fuel consumption
route_summary = data.groupby('route_id')['fuel_consumption'].mean().reset_index()

# Sort the data by fuel consumption in descending order
route_summary = route_summary.sort_values(by='fuel_consumption', ascending=False)

# Display the top 10 routes with the highest average fuel consumption
print("Top Routes by Average Fuel Consumption:")
display(route_summary.head(10))

# Visualize the average fuel consumption across routes
plt.figure(figsize=(12, 8))

sns.barplot(x='fuel_consumption', y='route_id', data=route_summary.head(10), palette='magma', hue=None)

plt.title('Top Routes by Average Fuel Consumption')
plt.xlabel('Average Fuel Consumption (liters)')
plt.ylabel('Route ID')
plt.tight_layout()
plt.show()
```

# Analyzing Relationship Between Engine Efficiency and Fuel Consumption

In this section, we aim to answer the third research question:
**"Is there a relationship between engine efficiency and fuel consumption?"**

We will:
1. Use a scatter plot to visualize the relationship between `engine_efficiency` and `fuel_consumption`.
2. Compute the correlation coefficient to measure the strength of the relationship.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Engine Efficiency vs. Fuel Consumption
plt.figure(figsize=(10, 6))
sns.scatterplot(x='engine_efficiency', y='fuel_consumption', data=data, color='blue', alpha=0.6)
plt.title('Relationship Between Engine Efficiency and Fuel Consumption')
plt.xlabel('Engine Efficiency (%)')
plt.ylabel('Fuel Consumption (liters)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate correlation coefficient
correlation = data['engine_efficiency'].corr(data['fuel_consumption'])
print(f"Correlation between Engine Efficiency and Fuel Consumption: {correlation:.2f}")
```

## Analysis of the Relationship Between Engine Efficiency and Fuel Consumption

### Observations:
1. The scatter plot shows **no clear relationship** between engine efficiency (%) and fuel consumption (liters).
2. The points are widely scattered with no visible trend or pattern, indicating that engine efficiency does not strongly influence fuel consumption in this dataset.

### Correlation:
- The correlation coefficient is **-0.03**, which is very close to zero.
- This confirms that there is **no significant linear correlation** between engine efficiency and fuel consumption.

### Conclusion:
The analysis suggests that fuel consumption is likely influenced by other factors (e.g., ship type, route, or weather conditions) rather than engine efficiency.


# Impact of Weather Conditions on Fuel Consumption and CO2 Emissions

In this section, we aim to answer the fourth research question:
**"What impact do weather conditions have on fuel consumption and CO2 emissions?"**

We will:
1. Group the data by `weather_conditions` to calculate average `fuel_consumption` and `CO2_emissions`.
2. Use bar plots to visualize the impact of different weather conditions on both metrics.


```python
# Group data by weather conditions and calculate averages for fuel consumption and CO2 emissions
weather_summary = data.groupby('weather_conditions')[['fuel_consumption', 'CO2_emissions']].mean().reset_index()

# Display summary table
print("Impact of Weather Conditions on Fuel Consumption and CO2 Emissions:")
display(weather_summary)

# Plot fuel consumption and CO2 emissions by weather conditions
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Average Fuel Consumption by Weather Condition
sns.barplot(x='weather_conditions', y='fuel_consumption', data=weather_summary, palette='Blues', ax=ax[0])
ax[0].set_title('Average Fuel Consumption by Weather Conditions')
ax[0].set_xlabel('Weather Conditions')
ax[0].set_ylabel('Fuel Consumption (liters)')

# Average CO2 Emissions by Weather Condition
sns.barplot(x='weather_conditions', y='CO2_emissions', data=weather_summary, palette='Oranges', ax=ax[1])
ax[1].set_title('Average CO2 Emissions by Weather Conditions')
ax[1].set_xlabel('Weather Conditions')
ax[1].set_ylabel('CO2 Emissions (kg)')

plt.tight_layout()
plt.show()
```

# CO2 Emissions Per Kilometer by Fuel Type

In this section, we aim to answer the fifth research question:
**"Which fuel type (HFO vs Diesel) has higher CO2 emissions per kilometer?"**

We will:
1. Calculate CO2 emissions per kilometer for each fuel type.
2. Compare the average CO2 emissions per kilometer between HFO and Diesel using a bar plot.


```python
# Create a new column for CO2 emissions per kilometer
data['CO2_per_km'] = data['CO2_emissions'] / data['distance']

# Group data by 'fuel_type' and calculate average CO2 emissions per kilometer
fuel_summary = data.groupby('fuel_type')['CO2_per_km'].mean().reset_index()

# Display the summary table
print("Average CO2 Emissions per Kilometer by Fuel Type:")
display(fuel_summary)

# Visualize CO2 emissions per kilometer by fuel type
plt.figure(figsize=(8, 5))
sns.barplot(x='fuel_type', y='CO2_per_km', data=fuel_summary, palette='coolwarm')
plt.title('Average CO2 Emissions per Kilometer by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('CO2 Emissions per Kilometer (kg/km)')
plt.tight_layout()
plt.show()
```

# Fuel Consumption Across Different Months

In this section, we aim to answer the sixth research question:
**"How does fuel consumption change over different months?"**

We will:
1. Group the data by `month` and calculate the average fuel consumption.
2. Use a line plot to visualize monthly trends in fuel consumption.


```python
# Group data by 'month' and calculate average fuel consumption
monthly_summary = data.groupby('month')['fuel_consumption'].mean().reset_index()

# Sort the months in calendar order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_summary['month'] = pd.Categorical(monthly_summary['month'], categories=month_order, ordered=True)
monthly_summary = monthly_summary.sort_values('month')

# Display the summary table
print("Average Fuel Consumption by Month:")
display(monthly_summary)

# Visualize average fuel consumption over months
plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='fuel_consumption', data=monthly_summary, marker='o', color='green')
plt.title('Average Fuel Consumption by Month')
plt.xlabel('Month')
plt.ylabel('Average Fuel Consumption (liters)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
